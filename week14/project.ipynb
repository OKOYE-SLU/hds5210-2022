```python
# I want to categorize the average cost of insulin in diffent regions
def categorize_avgcost(avgcost):
    if avgcost > 20:
        return 'expensive'
    elif avgcost < 10.0000:
        return 'cheap'
    else:
        return 'midpriced'
```

```python
#This function was created to generate the category of insulin price, we can alternate the price and get the category

categorize_avgcost(7), categorize_avgcost(16), categorize_avgcost(40)
```
